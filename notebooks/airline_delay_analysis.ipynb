```python
#immport necessary librarires
import pandas as pd
import numpy as np
```

### Search for Null values in the data

```python
df_airline.isnull().sum()
```

```python
# Clean the missing values
df_airline.fillna(0, inplace=True)
```

```python
columns_to_keep = [
    'year',
    'month',
    'carrier_name',
    'airport_name',
    'arr_flights',
    'arr_del15',
    'arr_delay',
    'carrier_delay',
    'weather_delay',
    'nas_delay',
    'security_delay',
    'late_aircraft_delay',
]
#Drop Unnecessary Columns
df_airline = df_airline[columns_to_keep]
df_airline.head()
```

```python
print(df_airline.columns.tolist())
```

###  1. Find which airports or airlines had the most delays between 2010 and 2025.

```python
airport_delays = df_airline.groupby('airport_name')[['arr_del15', 'arr_delay']].sum()
airport_delays
```

### Sort to find the Worst Airports

```python
airport_delays_sorted = airport_delays.sort_values(by='arr_del15', ascending=False)
```

```python
print(airport_delays_sorted.head(10))
```

### Delays by Airline(Carrier_name)

```python
#Group by Carrier Name
airline_delays = df_airline.groupby('carrier_name')[['arr_del15', 'arr_delay']].sum()
airline_delays.head()
```

```python
airline_delays_sorted = airline_delays.sort_values(by='arr_del15', ascending=False)
print(airline_delays_sorted.head(10))
```

“Southwest Airlines Co. and SkyWest Airlines Inc. lead in total delayed flights, which may negatively affect their Net Promoter Score and repeat business.”

### Normalize the Data

```python
# Delay rate = % of flights delayed
df_airline['delay_rate'] = df_airline['arr_del15'] / df_airline['arr_flights']
df_airline['delay_rate']
```

```python
# Average delay per delayed flight
df_airline['avg_delay_minutes'] = df_airline['arr_delay'] / df_airline['arr_del15']
df_airline['avg_delay_minutes']
```

### Visualize the Results

### 1. Top 10 Airlines by Total Delays

```python
import matplotlib.pyplot as plt

top_delays = airline_delays_sorted.head(10)
top_delays['arr_del15'].plot(kind='barh', figsize=(10,6), color='skyblue')
plt.title('Top 10 Airlines by Number of Delayed Flights')
plt.xlabel('Total Delayed Flights')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### 2. Average Delay Duration by Airline

```python
avg_delay_by_airline = df_airline.groupby('carrier_name')['avg_delay_minutes'].mean().sort_values(ascending=False).head(10)
avg_delay_by_airline.plot(kind='barh', figsize=(10,6), color='coral')
plt.title('Airlines with Highest Average Delay per Flight')
plt.xlabel('Avg. Delay Minutes')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### 3. Delay Rate by Airline

```python
import matplotlib.pyplot as plt

# Example for Delay Rate by Airline
ax = delay_rate_by_airline.plot(kind='barh', figsize=(10,6), color='green')
plt.title('Airlines with Highest Delay Rate')
plt.xlabel('Delay Rate (%)')
plt.gca().invert_yaxis()
for i, v in enumerate(delay_rate_by_airline):
    ax.text(v + 0.005, i, f"{v:.2%}", va='center')  # formats as percentage
plt.tight_layout()
plt.show()
```

### Analyze Delays (2010–2025)

```python
df_2010_25 = df_airline[(df_airline['year'] >= 2010) & (df_airline['year'] <= 2025)]
df_2010_25.head()
```

```python
yearly_delays = df_2010_25.groupby('year')[['arr_del15', 'arr_delay']].sum()
yearly_delays
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10,6))

# Plot delayed flights
ax1.plot(yearly_delays.index, yearly_delays['arr_del15'], marker='o', label='Delayed Flights', color='blue')
ax1.set_ylabel('Number of Delayed Flights', color='blue')
ax1.set_xlabel('Year')
ax1.tick_params(axis='y', labelcolor='blue')

# Create secondary axis for delay minutes
ax2 = ax1.twinx()
ax2.plot(yearly_delays.index, yearly_delays['arr_delay'], marker='o', label='Total Delay Minutes', color='red')
ax2.set_ylabel('Total Delay Minutes', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Flight Delays in U.S. (2010–2025)')
fig.tight_layout()
plt.grid(True)
plt.show()
```

✈️ Summary: 
U.S. Flight Delays (2010–2025)

From 2010 to 2019, the number of delayed flights and total delay minutes showed a gradual upward trend with minor fluctuations, indicating increasing air traffic and congestion.

In 2020, both metrics experienced a sharp decline due to the COVID-19 pandemic. The drastic drop in air travel led to significantly fewer flights and delays.

From 2021 onward, delays began to recover alongside the aviation industry's rebound. By 2023–2024, both the number of delayed flights and total delay minutes had surpassed pre-pandemic levels.

2025 shows a sudden drop, likely due to incomplete data (up to March only), and should be interpreted with caution.

### Plot 2019 vs 2020 Comparison

```python
df_2019 = df_2010_20[df_2010_20['year'] == 2019]
df_2020 = df_2010_20[df_2010_20['year'] == 2020]

metrics_2019 = {
    'Total Flights': df_2019['arr_flights'].sum(),
    'Delayed Flights': df_2019['arr_del15'].sum(),
    'Total Delay Minutes': df_2019['arr_delay'].sum()
}

metrics_2020 = {
    'Total Flights': df_2020['arr_flights'].sum(),
    'Delayed Flights': df_2020['arr_del15'].sum(),
    'Total Delay Minutes': df_2020['arr_delay'].sum()
}
```

### Create Comparison DataFrame

```python
comparison_df = pd.DataFrame([metrics_2019, metrics_2020], index=['2019', '2020'])
comparison_df
```

```python
comparison_df.plot(kind='bar', figsize=(10,6), color=['skyblue', 'salmon', 'lightgreen'])
plt.title('2019 vs 2020: Flights and Delays Comparison')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

✈️ 1. Significant Decline in Flight Volume

Total scheduled flights dropped from ~8.1 million in 2019 to ~5 million in 2020, a 38% decrease.
This reflects the COVID-19 pandemic's direct impact on air travel and mobility.

⚠️ 2. Steep Drop in Delayed Flights

Delayed flights fell from ~1.5 million to just ~469,000, a 69% decrease.
This is sharper than the drop in total flights, indicating that congestion was reduced and airports were operating below capacity.

⏱️ 3. Total Delay Time Shrinked

Total delay minutes dropped from ~106 million minutes in 2019 to ~29 million in 2020 — a 72% reduction.
This confirms fewer operational issues, possibly because:

There was less air traffic

Airlines had more turnaround flexibility

Fewer flights meant less chance of cascading delays

🧠 Final Business Insight:
"In 2020, the aviation industry experienced a sharp contraction in operations due to COVID-19, which led to a proportional reduction in both delayed flights and total delay time. The data suggests that reduced air traffic volume significantly alleviated operational congestion and delay occurrences."
